```python
import pandas as pd
from sklearn import datasets
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

```

```python
# Load sample dataset (Iris)
iris = datasets.load_iris()
data = iris.data

```

```python
# Define unsupervised learning algorithms
algorithms = {
    "KMeans": KMeans(n_clusters=3),
    "Agglomerative": AgglomerativeClustering(n_clusters=3),
    "DBSCAN": DBSCAN(eps=0.3, min_samples=10)
}

# Create a dictionary to store silhouette scores
silhouette_results = {}

```

```python
# Loop through each algorithm, fit and evaluate
for name, model in algorithms.items():
    model.fit(data)
    silhouette_results[name] = silhouette_score(data, model.labels_)

```

```python
# Print comparison results
print("Silhouette Scores for different Unsupervised Algorithms:")
for key, value in silhouette_results.items():
    print(f"{key}: {value}")

```

```python
# Optional: Visualize using PCA for dimensionality reduction
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(data)

```

```python
print(reduced_data)
```

```python
# Plot clusters for each algorithm
for name, model in algorithms.items():
    labels = model.fit_predict(data)
    plt.figure(figsize=(3, 3))
    plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=labels, cmap='viridis')
    plt.title(name)
    plt.show()
```

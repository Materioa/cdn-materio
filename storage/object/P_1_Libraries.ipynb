```python
import statistics as st
```

```python
a= [10, 20, 50, 10, 21, 90]
```

```python
mn = st.mean(a)
print("mean : ", mn)
```

```python
md = st.mode(a)
print("mode :", md)
```

```python
mdn = st.median(a)
print("median :", mdn)
```

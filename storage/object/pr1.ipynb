```python
import statistics as st
a = [10,12,30,45,60,75,90]
```

```python
mn = st.mean(a)
print("mean: ",mn)
```

```python
md = st.mode(a)
print("mode: ",md)
```

```python
mdn = st.median(a)
print("median: ",mdn)
```

```python
import numpy as np
b = np.array([[10,20,30,50,10], [1,2,3,5,6]])
print(b)
```

```python
print(b.ndim)
```

```python
b = np.array([[10,20,30,50,10], [1,2,3,5,6], [1,1,1,1,1]])
print(b)
print("Dim: ",b.ndim)
print("Shape: ",b.shape)
print("Size: ", b.size)
print("Item Size: ",b.itemsize) #integer size 4 or 8 acc. to complier
print("Data type: ",b.dtype)
```

```python
c = np.empty((5,5))
print(c)
```

```python
g = np.full((3,3),5)
print(g)
```

```python
d = np.arange(0,100,2) #(start, end, step) - create array with fixed interval
print(d)
```

```python
e = np.reshape(d, (10,5)) #reshape 'd' to 10 row 5 col
print(e)
```

```python
f = np.reshape(e, (2,5,5)) #reshape 'e' to 2 block's of 5x5
print(f)
print(f.ndim) 
```

```python
print("Last 10 elements: ",d[-10:])
```

```python
print("Alternative Numbers: ",d[::2])
```
